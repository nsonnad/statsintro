```python
from pylab import *
import scipy.stats as stats
import seaborn as sns
import pandas as pd
```

```python
x = randn(500)
```

```python
# scatterplot
plot(x, '.')
title('Scatterplot')
xlabel('X')
ylabel('Y')
```

```python
# histogram
hist(x)
xlabel('Data Values')
ylabel('Frequency')
title('Histogram, 25 bins')
```

```python
# cumulative probability density
numbins = 20
cdf = stats.cumfreq(x, numbins)
plot(cdf[0])
xlabel('Data Values')
ylabel('Cumulative Frequency')
title('Cumulative probability density function')
```

```python
# boxplot
# The error bars indiacte the range, and the box consists of the
# first, second (middle) and third quartile
boxplot(x)
title('Boxplot')
ylabel('Values')

# horizontal
boxplot(x, vert=False)
title('Boxplot, Horizontal')
xlabel('Values')
```

```python
# violin plot
nd = stats.norm
data = nd.rvs(size=(100))
nd2 = stats.norm(loc=0.5, scale=1.2)
data2 = nd2.rvs(size=(100))
```

```python
sns.set(context='poster')
df = pd.DataFrame({'Girls': data, 'Boys': data2})
sns.violinplot(df)
```

```python
# check for normality
_ = stats.probplot(x, plot=plt)
title('Problot - check for normality')
```

```python
# Bivariate data --------------------------
```

```python
# Generate data
x = randn(200)
y = 10 + 0.5 * x + randn(len(x))
```

```python
scatter(x, y)
```

```python
M = vstack((ones(len(x)), x)).T
pars = linalg.lstsq(M,y)[0]
intercept = pars[0]
slope = pars[1]
scatter(x,y)
hold(True)
plot(x, intercept + slope*x, 'r')
```
